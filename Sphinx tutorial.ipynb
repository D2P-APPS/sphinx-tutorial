## Generating documentation with docstrings
Sphinx can automatically turn code documentation into technical documentation.  

For this example, I've saved a module example_google/example_google.py in the project repository.  Sphinx will turn this module into technical documentation.

#### Install napoleon
    pip install sphinxcontrib-napoleon

#### Add the autodocumentation extension
In docs/conf.py, uncomment the following lines and adapt the path to your project -- here we're using an example from Google:

    import os
    import sys
    sys.path.insert(0, os.path.abspath('../example_google'))

Also in docs/conf.py, add napoleon to the extensions list and set it up to expect Google docstrings:  
    
    extensions = ['sphinx.ext.autodoc', 'sphinxcontrib.napoleon'
    ]
    napoleon_google_docstring = True

#### Generate documentation with sphinx-apidoc
From the docs/ folder, run

    sphinx-apidoc -f -o ./source ../example_google/

-f is to force the overwrite of existing files.  
-o is the relative path to the output directory ./source.  
../example_google is the relative path to the module directory.

This creates two new .rst files: source/modules.rst and source/example_google.rst.


#### Add modules the table of contents
In index.rst, add modules to the toc list. Note the entry 'source/modules' because modules.rst is in the source folder.

## Host your documentation on GitHub Pages

- Push your changes to GitHub
- In your GitHub repository, under Settings>GitHub Pages, set Source to "master branch /docs folder"
- Upon refresh, GitHub will display your site's URL
